```python
class WaterJugPuzzle():
  # modeling the initial state with two jugs of size 5 (liters) and 3 respectively, both start empty
  def __init__(self):
    self.five = 0
    self.three = 0

# checks for determining/modifying the status of the two jugs
  def is_five_empty(self):
    if self.five == 0: return True

  def is_three_empty(self):
    if self.three == 0: return True

  def is_five_full(self):
    if self.five == 5: return True

  def is_three_full(self):
    if self.three == 3: return True

  def state_ok(self):
    if self.five <= 5 and self.three <= 3: return True

  def save_state(self):
    self._five = self.five
    self._three = self.three

  def undo_state(self):
    self.five = self._five
    self.three = self._three

  ###################################################
  # Rule 1: 5L Jug is filled from nothing
  ###################################################

  def R1_fill_five(self):
    self.save_state()

    current = self.five
    difference = 5 - current
    self.five = self.five + difference

    if not self.state_ok(): self.undo_state()

  ###################################################
  # Rule 2: 3L Jug is filled from nothing
  ###################################################

  def R2_fill_three(self):
    self.save_state()

    current = self.three
    difference = 3 - current
    self.three = self.three + difference

    if not self.state_ok(): self.undo_state()

  ###################################################
  # Rule 3: 5L is filled from 3L
  ###################################################

  def R3_fill_five_from_three(self):
    self.save_state()

    fillable = 5 - self.five
    if self.three >= fillable:
      self.three = self.three - fillable
      self.five = self.five + fillable
    elif self.three > 0: # the above logic prevents situations where 5 is filled by 3, but not to the brim. this elif allows that
      fill = self.three
      self.three = self.three - fill
      self.five = self.five + fill

    if not self.state_ok(): self.undo_state()

  ###################################################
  # Rule 4: 3L is filled from 5L
  ###################################################

  def R4_fill_three_from_five(self):
    self.save_state()

    fillable = 3 - self.three
    if self.five >= fillable:
      self.five = self.five - fillable
      self.three = self.three + fillable
    elif self.five > 0:
      fill = self.five
      self.five = self.five - fill
      self.three = self.three + fill

    if not self.state_ok(): self.undo_state()

  ###################################################
  # Rule 5: Empty 5L
  ###################################################

  def R5_empty_five(self):
    self.save_state()

    self.five = 0

    if not self.state_ok(): self.undo_state()

  ###################################################
  # Rule 6: Empty 3L
  ###################################################

  def R6_empty_three(self):
    self.save_state()

    self.three = 0

    if not self.state_ok(): self.undo_state()



  def print_state(self): # printer function
    print(f"===> 5L: {self.five}  3L: {self.three}")
```

```python
wjp = WaterJugPuzzle()

# print initial state
wjp.print_state()

print("-----------------------") # apply a rule and print the resulting state
wjp.R1_fill_five()
wjp.print_state()

print("-----------------------")
wjp.R4_fill_three_from_five()
wjp.print_state()

print("-----------------------")
wjp.R6_empty_three()
wjp.print_state()

print("-----------------------")
wjp.R4_fill_three_from_five()
wjp.print_state()

print("-----------------------")
wjp.R1_fill_five()
wjp.print_state()

print("-----------------------")
wjp.R4_fill_three_from_five()
wjp.print_state()
```
